```python
#https://huggingface.co/suno/bark
#https://github.com/suno-ai/bark
#https://medium.com/@prateek.vyas/how-to-use-suno-bark-to-generate-text-to-speech-for-free-for-commercial-purpose-1e27ae2ba8d4
#https://medium.com/@vndee.huynh/build-your-own-voice-assistant-and-run-it-locally-whisper-ollama-bark-c80e6f815cba
from transformers import pipeline
import scipy

synthesiser = pipeline("text-to-speech", "suno/bark")
```

```python
speech = synthesiser("Hello, my dog is cooler than you!", forward_params={"do_sample": True})

#scipy.io.wavfile.write("bark_out.wav", rate=speech["sampling_rate"], data=speech["audio"])
```

```python
from IPython.display import Audio
#from bark import SAMPLE_RATE, generate_audio, preload_models

Audio(speech["audio"], rate=speech["sampling_rate"])
```

```python
speech = synthesiser("Bom dia, Morenino! Como está seu domingão???", forward_params={"do_sample": True})
Audio(speech["audio"], rate=speech["sampling_rate"])
```

```python
from pydub import AudioSegment

def convert_wav_to_mp3(input_path, output_path):
    audio = AudioSegment.from_wav(input_path)
    audio.export(output_path, format="mp3")

input_path = "bark_out.wav"
output_path = "bark_out.mp3"

convert_wav_to_mp3(input_path, output_path)
```
